```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, davies_bouldin_score
```

```python
dataset = pd.read_csv('iris.csv')

print('sample data:')
dataset.head()
```

```python
dataset.isna().sum()
```

```python
dataset.dtypes
```

```python
x = dataset.iloc[:, :4]
y = dataset['species']

xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.3, random_state=0 )
```

```python
dt = DecisionTreeClassifier()
dt.fit(xTrain, yTrain)
```

```python
print('Decision Tree Accuracy : {:.3f}'.format(accuracy_score(yTest, dt.predict(xTest))))
```

```python
plt.figure(figsize=(5, 5))
tree.plot_tree(dt, filled = True)
```

```python
KMeans = KMeans(n_clusters=3)
labels = KMeans.fit_predict(x)
```

```python
db_score = davies_bouldin_score(x, labels)
print(db_score)
```

```python
cols = dataset.columns
plt.scatter(x.loc[labels == 0, cols[0]], 
            x.loc[labels == 0, cols[1]], 
            s= 100, c='purple', 
            label = 'Setosa')
plt.scatter(x.loc[labels == 1, cols[0]], 
            x.loc[labels == 1, cols[1]], 
            s= 100, c='orange', 
            label = 'Versicolour')
plt.scatter(x.loc[labels == 2, cols[0]], 
            x.loc[labels == 2, cols[1]], 
            s= 100, c='green', 
            label = 'Virginica')

plt.scatter(KMeans.cluster_centers_[:, 0], 
            KMeans.cluster_centers_[:, 1], 
            s=100, c='red', 
            label='Centroids')

plt.legend()
```
